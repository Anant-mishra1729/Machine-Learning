## Missing Values

* Find features that have **NULL values**

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.heatmap(dataset.isnull(),cmap = 'viridis',cbar=False)
```

```python
'''
dataset.columns -> list of columns(str) in dataset
dataset[feature].isnull().sum() -> Returns count of True values (returned by isnull())
'''
fnull = []
print("Missing values")
for feature in dataset.columns:
    missing_count = dataset[feature].isnull().sum()
    if (missing_count):
        print(f"{feature} -> {missing_count} ({np.round(missing_count*100/dataset.shape[0],3)}%)")
        fnull.append(feature)
```

## Handling missing values

### *Find relationship between Missing values and Sales Price*

```python
for feature in fnull:
    data = dataset.copy()
    data[feature] = np.where(data[feature].isnull(),1,0)
```
